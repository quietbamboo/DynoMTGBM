```python
"""Refer and revise from MPEK https://github.com/kotori-y/mpek with MIT License"""
import random
import math
import pandas as pd
import matplotlib.pyplot as plt
import torch.utils.data as Data
from sklearn.preprocessing import StandardScaler
from copy import deepcopy
from torch import optim
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from scipy.stats import pearsonr
import sys
sys.path.insert(0, sys.path[0] + "/../")
from kinetic_params_evaluate.utils.mpek_utils import *
import matplotlib, os
matplotlib.rc("font", weight="bold")
import warnings
warnings.filterwarnings("ignore")

random_state = 66
random.seed(random_state)
np.random.seed(random_state)
torch.manual_seed(random_state)
torch.cuda.manual_seed(random_state)
torch.cuda.manual_seed_all(random_state)
current_dir = os.getcwd()


class MTLKineticParamDataset(Data.Dataset):
    def __init__(self, df, input_features, input_idx):
        self.df = df
        self.input_features = input_features
        self.input_idx = input_idx

    def __len__(self):
        return len(self.input_idx)

    def __getitem__(self, idx):
        idx_in_df = self.input_idx[idx]
        df_idx = self.df.loc[idx_in_df]

        label_km_values = np.array(df_idx[self.input_features[-3]])
        label_kcat_values = np.array(df_idx[self.input_features[-2]])
        label_kcatkm_values = np.array(df_idx[self.input_features[-1]])

        input_features_columns = self.input_features[:-3]  # features

        concatenated_series = [np.array(item) if isinstance(item, list) else np.array([item]) for item in
                               df_idx[input_features_columns]]
        input_values = np.concatenate(concatenated_series)

        return input_values, label_km_values, label_kcat_values, label_kcatkm_values


def mtl_return_loader(df_input, input_features, input_idx, batch_size, shuffle=True, seed=0):
    dataset = MTLKineticParamDataset(df_input, input_features, input_idx)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    g = torch.Generator()
    dataloader = Data.DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, generator=g)

    return dataloader


def return_scores(y_test, y_pred):
    # 移除 NaN 值
    mask = ~np.isnan(y_test)
    y_test_filtered = y_test[mask]
    y_pred_filtered = y_pred[mask]

    # scores
    rmse = np.sqrt(mean_squared_error(y_test_filtered, y_pred_filtered))
    mae = mean_absolute_error(y_test_filtered, y_pred_filtered)
    r2 = r2_score(y_test_filtered, y_pred_filtered)
    pcc = pearsonr(y_test_filtered, y_pred_filtered)

    return rmse, mae, r2, pcc[0]


def train_5cvs():
    scores_dict = {name: [] for name in score_names}
    val_scores_dict = deepcopy(scores_dict)
    task_scores_dict = {task_name: {'val': deepcopy(val_scores_dict)} for task_name in evaluate_task_list}

    for fold_idx, (fold_train_idx, fold_val_idx) in enumerate(df_fold_index.values.tolist(), start=1):
        train_one_cv(fold_train_idx, fold_val_idx, fold_idx, task_scores_dict)

    return task_scores_dict


def train_one_cv(fold_train_idx, fold_val_idx, fold_idx, task_scores_dict):
    # dataloader
    train_loader = mtl_return_loader(df_input, input_features, fold_train_idx, batch_size=params['batch_size'],
                                     shuffle=True, seed=random_state)
    val_loader = mtl_return_loader(df_input, input_features, fold_val_idx, batch_size=params['batch_size'],
                                   shuffle=True, seed=random_state)

    # model
    model = MPEKModel(expert_out=params['expert_out'], expert_hidden=params['expert_hidden'],
                      expert_layers=params['expert_layers'], num_experts=params['num_experts'],
                      num_tasks=len(task_dict), device=device).to(device)

    # optimizer
    pg = [p for p in model.parameters() if p.requires_grad]
    optimizer = optim.AdamW(pg, lr=params['lr'], weight_decay=5E-5)  # optimizer
    lf = lambda x: ((1 + math.cos(x * math.pi / params['epochs'])) / 2) * (1 - params['lrf']) + params['lrf']  # cosine
    scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lf)

    best_loss, best_model, best_val_scores = np.inf, None, None
    best_epoch, patience_nums = 0, 0

    train_losses = []
    val_losses = []

    for epoch_idx in range(params['epochs']):
        # train
        train_loss, model = train_one_epoch(model, optimizer, train_loader)
        scheduler.step()

        # evaluate
        val_loss, val_task_scores = evaluate_model(model, val_loader)
        train_loss_sum = np.sum(train_loss.cpu().numpy())
        val_loss_sum = np.sum(val_loss.cpu().numpy())

        # print Loss
        print(f"[Fold {fold_idx}] Ep {epoch_idx} Train MSE Loss {train_loss_sum:.5f} Val MSE Loss {val_loss_sum:.5f}")

        # compare
        if val_loss_sum <= best_loss:
            best_model = model
            best_epoch = epoch_idx
            best_val_scores = val_task_scores
            best_loss = val_loss_sum
            patience_nums = 0

        else:
            patience_nums += 1

        train_losses.append(train_loss)
        val_losses.append(val_loss)

        if patience_nums > params['patience']:
            print('Early stopped...')
            break

    # test and record scores
    print(f'[Fold {fold_idx}] Use model at epoch {best_epoch}')
    torch.save(best_model.state_dict(), f'./model_dict/{input_model}_fold{fold_idx}.pth')

    for task_name in evaluate_task_list:
        for score_idx, score_name in enumerate(score_names):
            task_scores_dict[task_name]['val'][score_name].append(best_val_scores[task_name][score_idx])

    print('Val', best_val_scores)
    # plot scores
    plot_log(train_losses, val_losses, fold_idx)


def print_scores(task_scores_dict):
    for task_name in evaluate_task_list:
        print(f"{task_name}\t RMSE\t MAE\t R2\t PCC\t")
        task_val_scores = task_scores_dict[task_name]['val']
        val_metrics = [f"{np.mean(task_val_scores[metric_name]):.4f}\t" for metric_name in score_names]
        print("Val  " + " ".join(val_metrics))
        print()


def plot_log(train_losses, val_losses, fold_idx):
    # Tensor to Numpy
    train_losses_np = [[tl_task.cpu().numpy() for tl_task in tl] for tl in train_losses]
    val_losses_np = [[vl_task.cpu().numpy() for vl_task in vl] for vl in val_losses]

    # plot
    plt.figure()
    epochs = range(1, len(train_losses) + 1)
    if len(task_dict) == 2:
        plt.plot(epochs, [tl[0] for tl in train_losses_np], label='Train log(Km) Loss')
        plt.plot(epochs, [tl[1] for tl in train_losses_np], label='Train log(kcat) Loss')
        plt.plot(epochs, [vl[0] for vl in val_losses_np], label='Val log(Km) Loss', linestyle='--')
        plt.plot(epochs, [vl[1] for vl in val_losses_np], label='Val log(kcat) Loss', linestyle='--')
    else:
        name = next(iter(task_dict))
        plt.plot(epochs, [tl[0] for tl in train_losses_np], label=f'Train {name} Loss')
        plt.plot(epochs, [vl[0] for vl in val_losses_np], label=f'Val {name} Loss', linestyle='--')

    plt.xlabel('$Epoch$', fontsize=12)
    plt.ylabel('$Loss$', fontsize=12)
    plt.title(f'Training and Validation Loss per Epoch Cross Validation {fold_idx}')
    plt.legend()
    plt.savefig(current_dir + f'/results/{input_model}_loss_plot_cv{fold_idx}.png')  # 保存图表为图片文件
    plt.show()


def train_one_epoch(model, optimizer, train_loader):
    model.train()
    accu_loss_train = torch.zeros(len(task_dict)).to(device)  # 累计损失
    valid_samples = torch.zeros(len(task_dict)).to(device)  # 有效样本计数

    optimizer.zero_grad()

    for step, data in enumerate(train_loader):
        data, label_km, label_kcat, label_kcatkm = [d.to(device).to(torch.float32) for d in data]
        preds = model(data)
        losses = _compute_loss(preds, [label_km, label_kcat], task_dict, device)

        # mask Nan
        valid = ~torch.isnan(losses)
        losses = torch.where(valid, losses, torch.tensor(0.0, device=device))

        final_loss = losses.sum()
        final_loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        accu_loss_train += losses.detach()

        # 计算每个任务的有效样本数
        for i in range(len(task_dict)):
            valid_samples[i] += valid[i].float().sum()

    accu_loss_train_mean = accu_loss_train / valid_samples

    return accu_loss_train_mean, model


def _compute_loss(preds, labels, task_dict, device):
    loss_fun = torch.nn.MSELoss()
    losses = torch.zeros(len(task_dict)).to(device)

    for tn, task in enumerate(task_dict.keys()):
        pred, label = preds[task], labels[tn]
        mask = ~torch.isnan(label)
        filtered_pred, filtered_label = pred[mask], label[mask]
        losses[tn] = loss_fun(filtered_pred, filtered_label.view(-1, 1))

    return losses


# def evaluate_model(model, data_loader):
#     model.eval()
#     all_pred = [[] for _ in range(len(evaluate_task_list))]
#     all_labels = [[] for _ in range(len(evaluate_task_list))]
#
#     accu_loss_eval = torch.zeros(len(task_dict)).to(device)  # 累计损失
#     valid_samples = torch.zeros(len(task_dict)).to(device)  # 有效样本计数
#
#     with torch.no_grad():
#         for step, data in enumerate(data_loader):
#             data, label_km, label_kcat = [d.to(device) for d in data]
#             preds = model(data)
#             losses = _compute_loss(preds, [label_km, label_kcat], task_dict, device)
#
#             valid = ~torch.isnan(losses)
#             losses = torch.where(valid, losses, torch.tensor(0.0, device=device))
#             accu_loss_eval += losses
#
#             for i in range(len(task_dict)):
#                 valid_samples[i] += valid[i].float().sum()
#
#             for i, task in enumerate(evaluate_task_list[:2]):
#                 all_pred[i].append(preds[task].cpu().numpy())
#                 all_labels[i].append([label_km.cpu().numpy(), label_kcat.cpu().numpy()][i])
#
#         for col_idx, col_name in enumerate(['logkm', 'logkcat']):
#             all_pred[col_idx] = scalers[col_name].inverse_transform(np.concatenate(all_pred[col_idx]))
#             all_labels[col_idx] = scalers[col_name].inverse_transform(np.concatenate(all_labels[col_idx]))
#
#         valid_mask = ~np.isnan(all_labels[0]) & ~np.isnan(all_labels[1])
#         all_pred[2] = (all_pred[1] - all_pred[0])[valid_mask]  # 预测的 log(kcat/km)
#         all_labels[2] = (all_labels[1] - all_labels[0])[valid_mask]  # log(kcat/km)
#
#     # cal scores
#     task_scores_results = {task_name: return_scores(np.concatenate(all_labels[idx]), np.concatenate(all_pred[idx]).ravel()) for idx, task_name in enumerate(evaluate_task_list)}
#
#     return accu_loss_eval / valid_samples, task_scores_results

def evaluate_model(model, data_loader):
    model.eval()
    all_pred = [[] for _ in range(len(evaluate_task_list))]
    all_labels = [[] for _ in range(len(evaluate_task_list))]

    accu_loss_eval = torch.zeros(len(task_dict)).to(device)
    valid_samples = torch.zeros(len(task_dict)).to(device)

    with torch.no_grad():
        for step, data in enumerate(data_loader):
            data, label_km, label_kcat, label_kcatkm = [d.to(device) for d in data]
            preds = model(data)
            losses = _compute_loss(preds, [label_km, label_kcat], task_dict, device)

            valid = ~torch.isnan(losses)
            losses = torch.where(valid, losses, torch.tensor(0.0, device=device))
            accu_loss_eval += losses
            valid_samples += valid.float()

            # 收集预测值和真实值（前两个任务：logkm 和 logkcat）
            all_pred[0].append(preds['km'].cpu().numpy())
            all_pred[1].append(preds['kcat'].cpu().numpy())
            all_labels[0].append(label_km.cpu().numpy())
            all_labels[1].append(label_kcat.cpu().numpy())
            all_labels[2].append(label_kcatkm.cpu().numpy())

        # 展平+反标准化
        for idx, task in enumerate(['logkm', 'logkcat']):
            all_pred[idx] = np.concatenate(all_pred[idx])
            all_labels[idx] = np.concatenate(all_labels[idx])

            all_pred[idx] = scalers[task].inverse_transform(all_pred[idx].reshape(-1, 1)).ravel()
            all_labels[idx] = scalers[task].inverse_transform(all_labels[idx].reshape(-1, 1)).ravel()
        all_labels[2] = np.concatenate(all_labels[2])
        all_labels[2] = scalers['logkcatkm'].inverse_transform(all_labels[2].reshape(-1, 1)).ravel()

        valid_mask = ~np.isnan(all_labels[2])
        all_labels[2] = all_labels[2][valid_mask]
        all_pred[2] = all_pred[1][valid_mask] - all_pred[0][valid_mask]

    # 计算得分
    task_scores_results = {
        task_name: return_scores(all_labels[idx], all_pred[idx])
        for idx, task_name in enumerate(evaluate_task_list)
    }

    return accu_loss_eval / valid_samples, task_scores_results


# TODO Config
input_model = 'MPEK'
params = {
    'expert_out': 768,
    'expert_hidden': 768,
    'expert_layers': 1,
    'num_experts': 4,
    'fold_num': 5,
    'lr': 1e-3,
    'lrf': 0.01,
    'epochs': 300,
    'batch_size': 2048,
    'patience': 30,
}
task_dict = {"km": 1, "kcat": 1}
evaluate_task_list = ["km", "kcat", 'kcatkm']
score_names = ['rmse', 'mae', 'r2', 'pcc']

# TODO Read dataset
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(f'Current mode with {input_model}. Device: {device}')

input_features = ['prott5', 'molebert', 'ph', 't', 'mw', 'logp', 'logkm', 'logkcat', 'logkcatkm']
print('Reading data...', end='')
df_input_path = current_dir + f'/../../data_process/dataset/df_standardized.pkl'
df_fold_index = pd.read_pickle(current_dir + f'/../../data_process/dataset/cdhit/cdhit_fold_index.pkl')
df_input = pd.read_pickle(df_input_path)
print('Finished.')

# standard labels
scalers = {}
scaled_labels = pd.DataFrame()
for col in ['logkm', 'logkcat', 'logkcatkm']:
    scaler = StandardScaler()
    scaled = scaler.fit_transform(df_input[[col]])
    df_input[col] = scaled.flatten()
    scalers[col] = scaler  # store

# TODO train
test_pred_list = []
task_scores_dict = train_5cvs()
print_scores(task_scores_dict)
```
