```python
import numpy as np
data = np.load('./results/pred_kcatkm_list_last.npy')
pred_logkcatkm_results = data[:, :, 2].mean(axis=0)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Arial'
import matplotlib
matplotlib.rc("font", family='Microsoft YaHei', weight="bold")

log_pred_value = pred_logkcatkm_results[:7]
experiment_df = pd.read_excel('./mutant_data.xlsx', sheet_name='IsTAL')
names = experiment_df['Number'].values
experiment_value = experiment_df['Experimental kcat/Km'].values
# 基准：野生型
wild_type_exp = experiment_value[0]
wild_type_pred = log_pred_value[0]

# 根据趋势一致性判断颜色
colors = []
for i in range(len(experiment_value)):
    if i == 0:
        colors.append("black")  # 野生型
    else:
        exp_trend = experiment_value[i] - wild_type_exp
        pred_trend = log_pred_value[i] - wild_type_pred
        if (exp_trend >= 0 and pred_trend >= 0) or (exp_trend < 0 and pred_trend < 0):
            colors.append("green")  # 正确预测
        else:
            colors.append("red")    # 错误预测

# 绘图
fig = plt.figure(dpi=300)
ax = fig.add_subplot(111)
# 设置网格在下方，并赋予透明度
ax.set_axisbelow(True)
plt.grid(True, alpha=0.5)
ax.scatter(names, log_pred_value, c=colors, s=88)

plt.xlabel('$Samples$', fontsize=14)
plt.ylabel('$Predicted$ $log(kcat/Km)$', fontsize=14)
plt.tick_params(axis="x", labelrotation=45, labelsize=12)
plt.tick_params(axis="y", labelsize=12)
plt.tight_layout()

#
# 可选：保存图像
plt.savefig("./results/IsTAL.png", dpi=300)
plt.show()
```
