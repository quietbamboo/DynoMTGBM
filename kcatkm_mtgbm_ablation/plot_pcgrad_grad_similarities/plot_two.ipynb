```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Arial'
import matplotlib, os
matplotlib.rc("font",family='MicroSoft YaHei',weight="bold")


def plot_gradient_similarity(gradient_similarities, label):
    if label == 'Default':
        line_style = '-'
    else:
        line_style = '--'
    for fold_idx in range(5):
        similarity_values = np.array(gradient_similarities[fold_idx])
        plt.plot([_ for _ in range(len(similarity_values))], similarity_values, label=f"{label} Fold {fold_idx + 1}",linewidth=2, linestyle=line_style)

default_gradient_similarities = np.load("default_gradient_similarities_log.npy")
pcgrad_gradient_similarities = np.load("PCGrad_gradient_similarities_log.npy")

fig, ax = plt.subplots(figsize=(8, 6))
plot_gradient_similarity(default_gradient_similarities, 'Default')
plot_gradient_similarity(pcgrad_gradient_similarities, 'PCGrad')

ax.set_xlabel('$Epochs$', fontsize=12)
ax.set_ylabel('$Gradient$ $Similarity$ $(Mean)$', fontsize=12)
plt.legend(loc='best')
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
plt.grid(False)
# plt.title('Mean Gradient Similarity Between Tasks Across Epochs')
plt.savefig('plot_two.png', dpi=300)
plt.show()
```
