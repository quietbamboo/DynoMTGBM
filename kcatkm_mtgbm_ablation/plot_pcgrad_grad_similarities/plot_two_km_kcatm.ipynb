```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Arial'
import matplotlib
matplotlib.rc("font",family='MicroSoft YaHei',weight="bold")
import os
current_dir = os.getcwd()

def plot_gradient_similarity(gradient_similarities):
    for fold_idx in range(5):
        similarity_values = np.array(gradient_similarities[fold_idx])
        plt.plot([_ for _ in range(len(similarity_values))], similarity_values, label=f"Fold {fold_idx + 1}",linewidth=1)

pcgrad_gradient_similarities = np.load("PCGrad_gradient_similarities_km_kcatm_log.npy")

fig = plt.figure(dpi=300)
ax = fig.add_subplot(111)
plot_gradient_similarity(pcgrad_gradient_similarities)

plt.xlabel('$Epochs$', fontsize=14)
plt.ylabel('$Gradient$ $Similarity$', fontsize=14)
plt.tick_params(axis="x", labelsize=14)
plt.tick_params(axis="y", labelsize=14)
plt.tight_layout()
plt.grid(False)
plt.legend(loc="best", fontsize=12)
plt.savefig(f"{current_dir}/plot_pcgrad.png", dpi=300)
plt.show()
```
