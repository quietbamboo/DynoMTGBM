```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Arial'
import matplotlib
matplotlib.rc("font", family='MicroSoft YaHei', weight="bold")

# 读取数据
df_mean_results = pd.read_excel('./results.xlsx')
model_names = df_mean_results['Model_name']
rmse_results = df_mean_results['Test_RMSE']
r2_results = df_mean_results['Test_R2']
pcc_results = df_mean_results['Test_PCC']
mae_results = df_mean_results['Test_MAE']

# TODO 绘制横向条形图 on test RMSE
fig = plt.figure(figsize=(6, 4), dpi=300)
ax = fig.add_subplot(111)
# 找出最好的模型（最低RMSE）
best_rmse_index = rmse_results.idxmin()
bars = ax.barh(model_names, rmse_results, edgecolor='black', height=0.6)  # Decrease height for thinner bars
# 设置最好的条形为橙色
bars[best_rmse_index].set_color('darkorange')
ax.set_xlabel("$Value$", fontsize=10)
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.grid(False)
plt.xlim([0, 4.5])
# Add text labels on top of the bars
for bar in bars:
    ax.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height() / 2,
            f'{bar.get_width():.3f}', va='center', fontsize=8)
plt.title("$RMSE$")
plt.tight_layout(pad=2.0)
# 保存并显示图像
plt.savefig("models_test_rmse_bar.png", dpi=300, bbox_inches="tight")
plt.show()

# TODO 绘制横向条形图 on test MAE
fig = plt.figure(figsize=(6, 4), dpi=300)
ax = fig.add_subplot(111)
# 找出最好的模型（最低MAE）
best_mae_index = mae_results.idxmin()
bars = ax.barh(model_names, mae_results, edgecolor='black', height=0.6)
# 设置最好的条形为橙色
bars[best_mae_index].set_color('darkorange')
ax.set_xlabel("$Value$", fontsize=10)
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.grid(False)
plt.xlim([0, 3.5])
# Add text labels on top of the bars
for bar in bars:
    ax.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height() / 2,
            f'{bar.get_width():.3f}', va='center', fontsize=8)
plt.title("$MAE$")
plt.tight_layout(pad=2.0)
# 保存并显示图像
plt.savefig("models_test_mae_bar.png", dpi=300, bbox_inches="tight")
plt.show()

# TODO 绘制横向条形图 on test PCC
fig = plt.figure(figsize=(6, 4), dpi=300)
ax = fig.add_subplot(111)
# 找出最好的模型（PCC最大）
best_pcc_index = pcc_results.idxmax()
bars = ax.barh(model_names, pcc_results, edgecolor='black', height=0.6)
# 设置最好的条形为橙色
bars[best_pcc_index].set_color('darkorange')
ax.set_xlabel("$Value$", fontsize=10)
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.grid(False)
plt.xlim([0, 0.84])
# Add text labels on top of the bars
for bar in bars:
    ax.text(bar.get_width()+0.01, bar.get_y() + bar.get_height() / 2,
            f'{bar.get_width():.3f}', va='center', fontsize=8)
plt.title("$PCC$")
plt.tight_layout(pad=2.0)
# 保存并显示图像
plt.savefig("models_test_pcc_bar.png", dpi=300, bbox_inches="tight")
plt.show()

# TODO 绘制横向条形图 on test R^2
fig = plt.figure(figsize=(6, 4), dpi=300)
ax = fig.add_subplot(111)
# 找出最好的模型（R^2最大）
best_r2_index = r2_results.idxmax()
bars = ax.barh(model_names, r2_results, edgecolor='black', height=0.6)
# 设置最好的条形为橙色
bars[best_r2_index].set_color('darkorange')
ax.set_xlabel("$Value$", fontsize=10)
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.grid(False)
plt.xlim([0, 0.63])
# Add text labels on top of the bars
for bar in bars:
    ax.text(bar.get_width() + 0.005, bar.get_y() + bar.get_height() / 2,
            f'{bar.get_width():.3f}', va='center', fontsize=8)
plt.title("$R^2$")
plt.tight_layout(pad=2.0)
# 保存并显示图像
plt.savefig("models_test_r2_bar.png", dpi=300, bbox_inches="tight")
plt.show()
```
